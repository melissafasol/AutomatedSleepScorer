## Function to splice data into 5 seconds epochs 

```python
dataset_splice_function = lambda dataset, t_len = 1252: [dataset[i:i + 1252] for i in range(0, (len(dataset)-1252), t_len)]
spliced_data = dataset_splice_function(data_4)
```

## Calculate amplitude threshold and if epoch exceeds threshold, then discard


```python
epochs_to_discard = []
slope_values = []
intercept_values = []
for idx, epoch in enumerate(spliced_data):
    ps = scipy.signal.welch(epoch, sampling_rate, window = 'hann', nperseg = nperseg)
    frequency = ps[0]
    power = ps[1]
    slope, intercept = np.polyfit(frequency, power, 1)
    if intercept > 1000 or slope < -1000:
        #print(idx)
        epochs_to_discard.append(idx)
        slope_values.append(slope)
        intercept_values.append(intercept)
    else:
        pass
```
