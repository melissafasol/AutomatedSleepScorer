```python
# datasets
import torch
from torch.utils.data import Dataset
from torch.utils.data import DataLoader

# neural networks
import torch.nn as nn
import torch.nn.functional as F

# optimizer (update weights - does backprop)
import torch.optim as optim

import matplotlib.pyplot as plt
```

## Load data

```python
# Dimensions: b x n x t x ... (b x t x n)
```

```python
class EEGDataset(Dataset):
    
    def __init__(self, n=1000, t_len=200):
        self._data = self._load_data(n, t_len)
    
    #function creates a dataset class
    def _load_data(self, n, t_len):
        data_list = []
        for i in range(n):
            noise = torch.normal(mean=0, std=0.1, size=(t_len,))
            signal = torch.sin(0.05*torch.arange(t_len) + 0.1 * i) 
            data_list.append(noise + signal)
        
        return torch.stack(data_list)
            
    def __len__(self):
        return len(self._data)
    
    def __getitem__(self, idx):
        return self._data[idx]
```

```python
eeg_dataset = EEGDataset()
```

```python
len(eeg_dataset[0])
```

```python
train_dataloader = DataLoader(eeg_dataset, batch_size= 64, shuffle=True)
```

## Create model

```python
class AutoMel(nn.Module):
    
    def __init__(self):
        super().__init__()
        feature_size = 20
        self.enc_lin = nn.Linear(200, feature_size)
        self.dec_lin = nn.Linear(feature_size, 200)
        
        nn.init.normal_(self.enc_lin.weight)
        nn.init.normal_(self.dec_lin.weight)
        
    def forward(self, x):
        # b x t
        features = self.enc_lin(x)#torch.sigmoid(self.enc_lin(x))
        
        decoding = self.dec_lin(features)
        
        return decoding
        
auto_mel = AutoMel()
```

## Train model

```python
criterion = nn.MSELoss()  # Defines the loss between the target and model prediction
optimizer = optim.Adam(auto_mel.parameters(), lr=0.001)  # Runs backprop and update weights

for epoch in range(1000):
    
    total_epoch_loss = 0
    
    for i, data in enumerate(train_dataloader):        
        optimizer.zero_grad() # zero the parameter gradients
        outputs = auto_mel(data)  # Generate our output
        loss = criterion(outputs, data)  # calcualte our loss
        loss.backward()  # calculate gradients
        total_epoch_loss = loss.item()
        optimizer.step()  # update weights using those gradients
    
    print(f"epoch {epoch} loss {total_epoch_loss}")
```
